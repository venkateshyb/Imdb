```python
import pandas as pd
import numpy as np
import matplotlib.pylab as mp
% matplotlib inline
df=pd.read_csv('imdb.csv')
df['cast']=df['cast'].astype(str)
idi=df.cast
df.info()
#print(type(idi))
```

```python
idi.dropna(inplace=True)
```

```python

lead=idi.copy()
lead1=idi.copy()
lead2=idi.copy()
lead3=idi.copy()
lead4=idi.copy()
for c in range(len(idi)-1):
    lead[c]=lead[c].split("|")[0]
    lead1[c]=idi[c].split("|")[1:2]
    lead2[c]=idi[c].split("|")[2:3]
    lead3[c]=idi[c].split("|")[3:4]
    lead4[c]=idi[c].split("|")[4:5]
#lead1=str(lead1)
lead=pd.DataFrame(data=lead)    
lead1=pd.DataFrame(data=lead1)
lead2=pd.DataFrame(data=lead2)
lead3=pd.DataFrame(data=lead3)
lead4=pd.DataFrame(data=lead4)
  
    
newrows=pd.merge(lead,lead1,left_index=True, right_index=True)
newrows1=pd.merge(lead2,lead3,left_index=True, right_index=True)
newrows2=pd.merge(newrows,newrows1,left_index=True, right_index=True)
df1=pd.merge(newrows2,lead4,left_index=True, right_index=True)
df1
```

```python
df1=df1.rename(columns={'cast_x_x':'Lead1','cast_y_x':'Lead2','cast_x_y':'Lead3','cast_y_y':'Lead4','cast':'Lead5'})
#df.drop(['cast'],axis=1,inplace=True)
df1.info()
```
