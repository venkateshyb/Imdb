```python
#I have taken all the necessary files for the database and have taken imdb.csv file as my input
import pandas as pd
import numpy as np
import matplotlib.pylab as mp
% matplotlib inline
df=pd.read_csv('imdb.csv')
df.info()
```

--Dropping all the tables which are not needed for the analysis.

```python
# Dropping the files with axis=1 and inplace True which changes your current table
df.drop(['id','imdb_id','homepage','tagline','keywords','overview','runtime','production_companies','release_date','vote_count','budget_adj','revenue_adj'],axis=1,inplace=True)
df.info()
```

--Removing All the duplicates from the dataset. Duplicates are not needed for your dataset

```python
# taking the sum of duplicates and dropping it from the current database
sum(df.duplicated())
df.drop_duplicates(inplace=True)
df.info()
```

--Dropping all the null values to normalize and make the table clean for Analysis

```python
#this will drop all the rows containing the null values from the database. Now the table is clean
df.dropna(inplace=True)
df.info()
```

--histogram for the current datase. This can help us in understanding the database for numerical values

```python
# hist is for analyzing all the numerical values and plotting it
df.hist(figsize=(10,8));
```

--From the above plotting we can understand that the budget and revenue are close to 0 to 0.5
There have been many values which have been initialized with zero. Popularity is varied based on the movie performance
release year and vote_average are looking good. With number of movies getting released is directly proportional to years.
The vote_average database is looking good

--I am dividing the cast column into number of Lead actors. So that i can perform my calculation with Lead Actors and 
popularity

```python
"""It was difficult to find this. Took me big time. Well we are using cast column to split the strings based on ' | '
character. Then we are using apply and lambda to create a small funtion that splits the strings and stores it as 
panda Series. we are doing this and saving the respective column as lead 1,2,3,4,5."""

foo = lambda x: pd.Series([i for i in (x.split('|'))])   
rev = df['cast'].apply(foo)
rev=rev.rename(columns={0:'Lead_1',1:'Lead_2',2:'Lead_3',3:'Lead_4',4:'Lead_5'})
```

-- We are merging the Rev table with df table and dropping the cast column since we dont need it any more.


```python
df1=pd.merge(df,rev,left_index=True, right_index=True)
df1.drop(['cast'],axis=1,inplace=True)
df1.info()
```

```python
Checking for the rows which has missing values . Now i have to take them out.
```

```python
df1[df1.Lead_2.isnull()]
```

```python
# Since Lead1 doesn't have any missing values. I am going with it. I dont want these 4 missing rows.
df1.drop(['Lead_2','Lead_3','Lead_4','Lead_5'],axis=1,inplace=True)
```

```python
#ready for my Analysis
df1.info()
```

```python
1.Does the revenue of the movie depends upon its popularity?
```

```python
# I am copying the original file to news so that it doesn't damage my database
#i am finding pop to get the popularity more than the mean
#Vice versa for n pop
#I am using histogram to plot the revenue and popularity.
news=df1.copy()
pop=(news.popularity)>(news.popularity.mean())
npop=(news.popularity)<(news.popularity.mean())
news.revenue[pop].hist(alpha=0.5,label='popular')
news.revenue[npop].hist(alpha=0.5,label='notpopular')
mp.xlabel("Revenue",fontsize=10)
mp.ylabel("Frequency",fontsize=10)
mp.title("Popularity Vs Revenue")
mp.legend();
```

```python
Yes the revenue of the movie does depend on its popularity. From this graph we can see that whenever the popularity
increases the revenue increases. So inorder to make the movie successfull we have to make it popular.
```

```python
2.Does the popular movies have more vote_average than the unpopular movies?
```

```python
# we are using the same pop and npop to find the popularity based on vote_average?
#we are grouping the values based on vote_average and popularity
#I am using histogram to plot the vote_average and popularity.
pop=(news.popularity)>(news.popularity.mean())
npop=(news.popularity)<(news.popularity.mean())
news.vote_average[pop].hist(alpha=0.5,label='popular')
news.vote_average[npop].hist(alpha=0.5,label='notpopular')
mp.xlabel("Vote_Average",fontsize=10)
mp.ylabel("Frequency",fontsize=10)
mp.title("Vote_Average Vs Popularity")
mp.legend();
```

OMG!! From this we can see that the unpopular movies has higher vote average than the popular movies. 
I did not expect this. Some Unpopular were quite successfull in votes than the popular movies. Only it was not popular. 
That is interesting

```python
 3.Top 5 directors based on vote average?
```

```python
#I am trying to find the top 5 directors based on vote_average
# I am grouping by directors and mean of vote average to find the result
#i am using n largest to find the top 5 rows.
# I dont know how to name the y axis if u have any idea let me know

counting=news.groupby('director').vote_average.mean()
#counting
plot1=counting.nlargest(5).plot(kind='bar',title='Above_average_directors')
plot1.set(xlabel="Directors", ylabel="Mean_Count");
```

```python
4.Top 5 directors who have produced highest profits in there movies?
```

```python
# i am trying to groupby directors and create a new column called difference to find the 
#difference between the revenue and budget. And then plotting the bar graph for the same questionm
news['diff'] =news.revenue-news.budget
counting=news.groupby('director').diff.mean()
plot1=counting.nlargest(5).plot(kind='bar',title='High_margin_directors')
plot1.set(xlabel="Directors", ylabel="Mean_Profits");

```

```python
# I am using the data to group directors based on budget and revenue.
# i am grouping the director and finding the mean for the revenue and budget column
counting=news.groupby('director').revenue.mean()
counting1=news.groupby('director').budget.mean()
# Using the bar chart to plot the graph for top 5 directors based on revenue and budget
counting.nlargest(5).plot(kind='bar',alpha=0.5,color='green',label='revenue')
counting1.nlargest(5).plot(kind='bar',alpha=0.5,color='blue',label='budget')
mp.title('Directors with Mean Revenue and Budget')
mp.ylabel('Budget and Revenue')
mp.legend();
```

```python
6.How does the popularity for movie increases based on last 15 years?
```

```python
# From this i have splitted the year of releasing after 2000 since there is too much data
# The data is then gouped by release year and mean of popularity using the bar chart. 

latest=news.query('release_year>2000')
latestpop1=latest.groupby('release_year').popularity.mean().plot(kind='bar',title='Popular_years')
latestpop1.set(ylabel="Mean_Popularity");
```

From this chart we can see that the popularity for movies increases as year increases. For the last 3 years there 
have been some very good movies. And this count is going to keep increasing.

```python
7.Top 5 directors based on popularity?
```

```python
#Since popularity is our dependent variable. These are our popular directors based on popularity
#we are grouping by directors and finding the mean of the values.
counting=news.groupby('director').popularity.mean()
#counting
plot1=counting.nlargest(5).plot(kind='bar',title='Popular Directors')
plot1.set(xlabel="Directors", ylabel="Mean_Popularity");
```

```python
8. How does the popularity differ based on Lead Actors?
```

```python
#Based on the lead actors we split above we have grouped by Lead1 and the popularity mean to find the
#largest 5 lead actors.
counting=news.groupby('Lead_1').popularity.mean()
#counting
plot1=counting.nlargest(5).plot(kind='bar',title='Popular Actors');
plot1.set(xlabel="Lead Actors", ylabel="Mean_Popularity");
```

Chris Pratt is going on a roll. He is making a impact with his movies. He has huge difference from others. 
More than double the popularity from the 2nd most

```python
9.How does the budget and revenue changes based on last 15 years?
```

```python
# We are finding the budget for the movies based on years. 
#The movies is taken above the year 2000 to calculate it
# the data is then grouped by year and the mean budget is taken and plotted using bar chart.
latest=news.query('release_year>2000')
plot1=latest.groupby('release_year').budget.mean().plot(kind='bar',title='Years based on budget')
plot1.set(xlabel="Years", ylabel="Mean Budget");
```

From the graph we can see that the budget is decreasing based on years. Thats quite bad and it must be due to 0 values.

```python
plot1=latest.groupby('release_year').revenue.mean().plot(kind='bar',title='Years based on Revenue')
plot1.set(xlabel="Years", ylabel="Mean Revenue");
```

However we can see that the revenue is increasing irrespective of the budget. Some low budget movies are doing
fine in the box office collection.There has been ups and down in revenue and budget. But from this we can see that the budget is been reducing for the last five years, but the revenue is increasing irrespective of budget.
